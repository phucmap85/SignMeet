# Load dict

```python
import json
import os
import shutil
```

```python
df = json.load(open('WLASL_v0.3.json'))
```

```python
word_id = {}
```

```python
for line in open('wlasl_class_list.txt').readlines():
    ou = line.strip().split('\t')
    word_id[ou[1]] = int(ou[0])
```

```python
word_id['bad']
```

# Copy videos to other folder

```python
wordlist = {'hello', 'you', 'me', 'thank you', 'and', 'name', 'tomorrow', 'yesterday', 'today', 'what', 'where', 'how', 'when', 
            'who', 'why', 'fine', 'bad'}
```

```python
for word in wordlist:
    os.mkdir(os.path.join('data', word))
    for i in range(len(df[word_id[word]]['instances'])):
        video_name = str(df[word_id[word]]['instances'][i]['video_id'] + '.mp4')
        video_path = 'videos/' + video_name
        if os.path.isfile(video_path):
            shutil.copy(video_path, 'data/' + word)
    print('\n')
```
